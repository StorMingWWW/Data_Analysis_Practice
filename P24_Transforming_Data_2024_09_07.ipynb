### DataFrame的分组、聚合操作

```python
S_01_01 = pd.Series(["001", "002", "001", "002", 
                     "001", "002", "001", "002"])
S_02_01 = pd.Series(["2022Q1", "2022Q1", "2022Q1", "2022Q1", 
                     "2022Q2", "2022Q2", "2022Q2", "2022Q2"])
S_03_01 = pd.Series(["生鲜食品", "生鲜食品", "休闲食品", "休闲食品", 
                     "生鲜食品", "生鲜食品", "休闲食品", "休闲食品"])
S_04_01 = pd.Series([1500, 2000, 3000, 2500, 
                     1800, 2200, 3200, 2700])
S_05_01 = pd.Series([105, 84, 171, 162, 67, 150, 99, 57])
DF1 = pd.DataFrame({"分店编号": S_01_01, "时间段": S_02_01, 
                    "商品类别": S_03_01, "销售额": S_04_01,
                    "销售数量": S_05_01})
```

```python
DF1
```

对DataFrame进行分组运算

```python
# 把 DF1 针对分店编号进行分组， 就算销售额平均值
DF1.groupby("分店编号")["销售额"].mean()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;聚合函数<br>
`count()` 得到元素的数量&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;`min()` 得到所有元素的最小值<br>
`first()` 得到第一个元素&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;`max()` 得到所有元素的最大值<br>
`last()` 得到最后一个元素&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;`std()` 得到所有元素的标准差<br>
`mean()` 得到所有元素的平均值&nbsp;&nbsp;&nbsp;&nbsp;`var()` 得到所有元素的方差<br>
`median()` 得到所有元素的中位数&nbsp;&nbsp;&nbsp;&nbsp;`prod()` 得到所有元素的积<br>
`sum()` 得到所有元素的和

```python
DF1.groupby("分店编号")[["销售额", "销售数量"]].mean()
```

```python
# 把 DF1 针对分店编号和时间段进行分组 ，计算销售额和销售数量的平均值
DF1.groupby(["分店编号", "时间段"])[["销售额", "销售数量"]].mean()
```

```python
def max_plus_10(nums):
    return nums.max() + 10
```

```python
# 把 DF1 针对分店编号和时间段进行分组， 对销售额和销售数量进行 max_plus_10 操作
DF1.groupby(["分店编号", "时间段"])[["销售额", "销售数量"]].apply(max_plus_10)
```

使用透视表<br>
`pd.pivot_table`

```python
# 把 DF1 的分店编号和时间段作为索引， 商品类别作为列， 计算销售额的总和
pd.pivot_table(DF1, index=["分店编号", "时间段"], columns="商品类别", values="销售额", aggfunc=np.sum)
```

```python
# 把 DF1 的分店编号和时间段作为索引， 商品类别作为列， 计算销售额的平均值
pd.pivot_table(DF1, index=["分店编号", "时间段"], columns="商品类别", values="销售额")
```

```python
# 把 DF1 的商品类别作为索引， 分店编号作为列， 计算销售额的平均值
pd.pivot_table(DF1, index = "商品类别", columns="分店编号", values="销售额", aggfunc=np.mean) 
```

```python
pd.pivot_table(DF1, index = "商品类别", columns="分店编号", values="销售额", aggfunc=np.sum) 
```

```python
# 1. 定义年龄分组列表: [0, 10, 20, 30, 40, 50, 60, 120]
# 2. 并根据以上分组对 DF2 的年龄列进行分组
Age_Bins = [0, 10, 20, 30, 40, 50, 60, 120]
pd.cut(DF2["年龄"], Age_Bins)
```

```python
# 定义分组标签列表: ['儿童', '青少年', '青年', '壮年', '中年', '中老年', '老年']
# 根据前一个单元格里定义的分组对 DF2 的年龄列进行分箱， 并使用以上分组标签
# 最后为 DF2 新建 "年龄组" 列， 值为以上分组标签
Age_Labels = ['儿童', '青少年', '青年', '壮年', '中年', '中老年', '老年']
DF2["年龄组"] = pd.cut(DF2["年龄"], Age_Bins, labels= Age_Labels)
DF2
```

```python
# 对 DF2 根据年龄组进行分配， 计算各个年龄组的平均工资
DF2.groupby("年龄组")["工资"].mean()
```
