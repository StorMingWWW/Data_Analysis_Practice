```python
import numpy as np
import pandas as pd
```

重命名索引和列名

```python
S_ID1 = pd.Series(["孙哥", "黄哥", "翘兔", "sed老师", "狗处", "死矮子"])
S_Gender1 = pd.Series(["女", "男", "男", "女", "女", "女"])
S_Age1 = pd.Series([58, 27, 32, 58, 27, 51])
S_E_mail1 = pd.Series(["sunge@randommail.com",
                       "huangge@inboxrandom.com", 
                       "qiaotu@tempmail.org", 
                       "mastersed@randomize.com", 
                       "rubbishgou@fakebox.net", 
                       "shorty@tempinbox.com"])
DF1 = pd.DataFrame({"客户_姓名":S_ID1 , "客户 性别":S_Gender1 , 
                    "Age":S_Age1 , "邮箱":S_E_mail1})
New_Index1 = ["1", "2_", "3", "4", "_5", "6*"]
DF1.index = New_Index1

DF1
```

```python
DF1.rename(index = {"2_":"2", "_5":"5", "6*":"6"})
```

```python
DF1.rename(columns = {"客户_姓名":"客户姓名", "客户 性别":"客户性别", 
                      "Age":"客户年龄", "邮箱":"客户邮箱"})
```

```python
DF1 = DF1.rename(index = {"2_":"2", "_5":"5", "6*":"6"})
DF1
```

```python
DF1.rename(columns = {"客户_姓名":"客户姓名", "客户 性别":"客户性别", 
                      "Age":"客户年龄", "邮箱":"客户邮箱"}, inplace = True)
DF1
```
