创建一个名为`Students`的`DataFrame`包含数据如下：

`{"001":{"姓名": "孙哥", "考试1": 85, "考试2": 95, "考试3":92}, 
"002":{"姓名": "黄哥", "考试1": 91, "考试2": 92, "考试3":94}, 
"003":{"姓名": "翘兔", "考试1": 86, "考试2": 81, "考试3":89}, 
"004":{"姓名": "sed老师", "考试1": 79, "考试2": 89, "考试3":95}, 
"005":{"姓名": "狗处", "考试1": 96, "考试2": 91, "考试3":91}, 
"006":{"姓名": "死矮子", "考试1": 81, "考试2": 89, "考试3":92} }`

要求索引为学号 {001、 002等}， 列名为姓名、 考试1、 考试2、 考试3， 并输出创建好的DataFrame。

```python
Student_Data = {"001":{"姓名": "孙哥", "考试1": 85, "考试2": 95, "考试3":92}, 
"002":{"姓名": "黄哥", "考试1": 91, "考试2": 92, "考试3":94}, 
"003":{"姓名": "翘兔", "考试1": 86, "考试2": 81, "考试3":89}, 
"004":{"姓名": "sed老师", "考试1": 79, "考试2": 89, "考试3":95}, 
"005":{"姓名": "狗处", "考试1": 96, "考试2": 91, "考试3":91}, 
"006":{"姓名": "死矮子", "考试1": 81, "考试2": 89, "考试3":92} }
Students = pd.DataFrame(Student_Data).T
Students
```

给`Students`增加一列`考试4`，数据依次为`[72, 69, 79, 83, 82, 76]`。

```python
# Students["考试4"] = [72, 69, 79, 83, 82, 76]
Students["考试4"] = pd.Series([72, 69, 79, 83, 82, 76],
                            index = ["001", "002", "003", "004", "005", "006"])
Students
```

给`Students`添加一行，索引为 007 ， 数据为`"小杨": 79, 82, 81, 69`。

```python
Students. loc["007"] = ["钱哥", 79, 82, 81, 69]
Students
```

输出`Students`删除索引为 006、 007 的结果，但不要改变`Students`本身。

```python
Students.drop(["006", "007"], axis = 0)
```
