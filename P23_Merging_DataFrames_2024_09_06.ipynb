```python
S_01_16 = pd.Series(["1001", "1002", "1003", "1004", "1005", "1006"])
S_02_16 = pd.Series(["2000-03-12", "2000-03-13", 
                     "2000-03-13", "2000-03-15", 
                     "2000-03-18", "2000-03-21"])
S_03_16 = pd.Series(["1", "1", "2", "6", "5", "3"])
S_04_16 = pd.Series(["A", "B", "C", "D", "E", "F"])
S_05_16 = pd.Series([2,3, 1, 4, 5, 2])
Order_DF4 = pd.DataFrame({"订单ID": S_01_16, "订单日期": S_02_16, 
                          "客户ID": S_03_16, "产品": S_04_16,
                          "数量": S_05_16})
Order_DF4
```

```python
# 根据索引合并 Customer_DF4 和 Order_DF4 ，列后缀分别为 "_Customer" 和 "_Order"
Customer_DF4.join(Order_DF4, how = "inner", lsuffix="_Customer", rsuffix="_Order")
```
